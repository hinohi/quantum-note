```python
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
def make_mod(n: int, a: int) -> list[int]:
    data = [1]
    while True:
        r = (data[-1] * a) % n
        if r == 1:
            return data
        data.append(r)
```

```python
11 * 13
```

```python
data = make_mod(141, 2)
x = np.arange(2 ** 10)
y = []
for i in range(len(x)):
    y.append(data[i % len(data)])
```

```python
fig, (a1, a2) = plt.subplots(2, 1, figsize=(15, 5))

a1.plot(x, y)
a1.grid()

yi = np.fft.fft(y)
a2.bar(x, np.abs(yi))
a2.grid()
```

```python
x[np.abs(yi) > 6000]
```

```python
math.gcd(2 ** (46 // 2) + 1, 141)
```

```python
math.gcd(2 ** (46 // 2) - 1, 141)
```
